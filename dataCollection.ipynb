```python
topics = [
    "Automobile",
    "Car classification",
    "Automotive industry",
    "History of the automobile",
    "Internal combustion engine",
    "Electric vehicle",
    "Hybrid vehicle",
    "Transmission (mechanics)",
    "Brake",
    "Suspension (vehicle)",
    "Car safety",
    "Autonomous car",
    "Vehicle emissions control",
    "Fuel economy in automobiles",
    "Automotive design"
]
```

```python
import wikipediaapi, time, os

wiki = wikipediaapi.Wikipedia(user_agent='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36', language='en')

os.makedirs("data", exist_ok=True)

for topic in topics[2:]:
    page = wiki.page(topic)
    print(f"on {topic}")
    if page.exists():
        with open(f"data/{topic.replace(' ', '_')}.txt", "w", encoding="utf-8") as f:
            f.write(page.text)
    time.sleep(5)
```
